```python
import matplotlib.pyplot as plt

# Sample data (replace this with your data)
# final_recap_df = ...  # Your DataFrame containing 'scenario', 'confidence_avg', and 'dataset' columns

# Get unique scenarios and datasets
scenarios = final_recap_df['scenario'].unique()
datasets = final_recap_df['dataset'].unique()
num_scenarios = len(scenarios)
num_datasets = len(datasets)

# Create violin plots using matplotlib
plt.figure(figsize=(10, 6))
for i, dataset in enumerate(datasets):
    dataset_data = final_recap_df[final_recap_df['dataset'] == dataset]
    positions = [j + i * 0.4 - 0.2 * (num_datasets - 1) for j in range(num_scenarios)]
    for j, scenario in enumerate(scenarios):
        scenario_data = dataset_data[dataset_data['scenario'] == scenario]['confidence_avg']
        plt.violinplot([scenario_data], positions=[positions[j]], widths=0.3, showmedians=True, showextrema=True)

# Set labels and title
plt.xlabel('Scenario')
plt.ylabel('Average Confidence')
plt.title('Violin Plot of Average Confidence by Scenario')

# Set x-axis ticks and labels
plt.xticks(range(num_scenarios), scenarios)

# Add legend
plt.legend(datasets, title='Dataset', loc='lower right')

plt.grid(True)
plt.show()
plt.close()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame (Replace this with your DataFrame)
# data = {
#     'encoder_type': ['Encoder A', 'Encoder A', 'Encoder A', 'Encoder B', 'Encoder B', 'Encoder B', 'Encoder C', 'Encoder C', 'Encoder C'],
#     'loss_fc': ['Loss1', 'Loss2', 'Loss3', 'Loss1', 'Loss2', 'Loss3', 'Loss1', 'Loss2', 'Loss3'],
#     'accuracy': [0.8, 0.85, 0.9, 0.75, 0.78, 0.82, 0.88, 0.86, 0.84],
#     'precision': [0.7, 0.75, 0.8, 0.65, 0.68, 0.72, 0.78, 0.76, 0.74]
# }
data_path = os.path.join('experiments', 'recap', 'recap-best-all.xlsx')
df = pd.read_excel(data_path)
df = df[df['dataset'] == 'filtered']
# Find the best performing model for each combination of encoder type and loss function
best_models = df.groupby(['encoder', 'scenario']).apply(lambda x: x.loc[x['accuracy'].idxmax()])
# Plot pairwise comparison using scatter plots
plt.figure(figsize=(5, 3))
sns.scatterplot(data=best_models, x='accuracy', y='f1_score', hue='encoder', style='scenario', s=100)

# Set labels and title
plt.xlabel('Accuracy')
plt.ylabel('F1-Score')
plt.title('Pairwise Comparison of Best Performing Models')

plt.legend(title='Legend', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame (Replace this with your DataFrame)
data_path = os.path.join('experiments', 'recap', 'recap-best-all.xlsx')
df = pd.read_excel(data_path)
df = df[df['dataset'] == 'filtered']
# Melt the DataFrame to have metrics as a variable
# df_melted = pd.melt(df, id_vars='scenario', var_name='metric', value_name='value')
# print(df_melted)
# List of encoder types
encoder_types = df['encoder'].unique()

# Plot each encoder type separately
for encoder_type in encoder_types:
    # Filter the DataFrame for the current encoder type
    df_filtered = df[df['encoder'] == encoder_type]
    df_filtered = df_filtered[['scenario', 'accuracy', 'precision', 'recall', 'f1_score']]
    df_filtered = pd.melt(df_filtered, id_vars='scenario', var_name='metric', value_name='value')
    # Plot
    plt.figure(figsize=(12, 8))
    sns.barplot(data=df_filtered, x='scenario', y='value', hue='metric', ci=None, hue_order=['accuracy', 'precision', 'recall', 'f1_score'])

    # Set labels and title
    plt.xlabel('Loss Function')
    plt.ylabel('Metric Value')
    plt.title(f'Comparison of Metrics for {encoder_type} by Loss Function')

    plt.legend(title='Metric', bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.xticks(rotation=45, ha='right')

    plt.grid(axis='y')
    plt.show()
```
