```python
import pandas as pd

batch_size_df = pd.read_excel('experiments/recap/batch_size/recap_overall_batch_size.xlsx')
batch_size_df['confidence_avg'] = batch_size_df['confidence_avg'] * 100
batch_size_df
```

```python
filtered_df = batch_size_df[batch_size_df['dataset'] == 'filtered']
filtered_df = filtered_df[filtered_df['encoder'] == 'lstm']
filtered_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame
# data = {'Batch Size': [4, 8, 16, 32, 64, 128],
#         'Accuracy': [0.85, 0.88, 0.87, 0.90, 0.91, 0.92],
#         'F1 Score': [0.75, 0.78, 0.77, 0.80, 0.82, 0.83]}
# df = pd.DataFrame(data)
plt.figure(figsize=(5, 2.5))

# Plotting
bar_width = 0.8
index = filtered_df.index
# light to dark
# 99DDC8
# 95BF74
# 659B5E
# 556F44
# 283F3B
plt.bar(index, filtered_df['accuracy'], bar_width, label='Accuracy')
plt.bar(index + bar_width, filtered_df['precision'], bar_width, label='Pre')
plt.bar(index + bar_width + bar_width, filtered_df['recall'], bar_width, label='Recall')
plt.bar(index + bar_width + bar_width + bar_width, filtered_df['f1_score'], bar_width, label='F1')
# dark to light
# 493B2A
# 593F62
# 7B6D8D
# 8499B1
# A5C4D4
# Adding labels and title
plt.xlabel('Batch Size')
plt.ylabel('Score')
# plt.title('Accuracy and F1 Score by Batch Size')
plt.xticks(index + bar_width, filtered_df['batch_size'])
plt.legend(loc='upper right')
plt.ylim(60, 100)
# Show plot
plt.tight_layout()
plt.grid(True)
plt.savefig(os.path.join('visualization', "batch_size_fr.pdf"), bbox_inches='tight')

plt.show()
plt.close()
```
